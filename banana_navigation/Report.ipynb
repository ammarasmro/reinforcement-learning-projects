# Methodology

## Algorithm: DQN

To solve this task, we used Deep Q-Networks. This deep network uses a neural networks to learn action value functions. So at each timestep, given the state of the environment, whether from gym variables or directly from pixels, the neural net will output a value for each action.

To play a role using the algorithm, the agent needs only to pick the action with the maximum output value. 

## Model

For the model I used the DQN algorithm to solve this task. 

**Model Architecture**
1. Linear layer | shape(in = 37, out = 64)
2. ReLU activation layer
3. Linear layer | shape(in = 64, out = 64)
4. ReLU activation layer
5. Linear layer | shape(in = 64, out = 4)

As my first DeepRL project, I actually thought that I will need a much bigger model. Through experiments and multiple runs I found out that even three layers with such a low number of hidden units did the job. 

**Hyper Parameters**
```
BUFFER_SIZE = 10000 sample  # replay buffer size
BATCH_SIZE = 64             # minibatch size
GAMMA = 0.99                # discount factor
TAU = 1e-3                  # for soft update of target parameters
LR = 5e-4                   # learning rate 
UPDATE_EVERY = 4            # how often to update the network
```
We didn't need to do alot of find tuning for this task as it was solved easily. Varying the learning rate might be benefitial for future experiments.
